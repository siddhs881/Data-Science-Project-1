**Data Cleaning**

**Cleaning Actions Performed:**
1. Converted very long survey questions into short variable names.
2. Dropped columns irrelevant to my 5 specific research questions
3. The 'degree_program' column contained messy entries. The entries were unified to specific computing tracks.
4. Students who selected mutliple options in the 'race_ethnicity' column were mapped to a single "Multi-Racial" category.

**Citations:**
- Used Github Copilot to generate the 'columns_to_keep' list based on my research questions and condensing major lists.

```python
# Using Copilot to keep the columns needed
# Keep only the columns needed for our specific research questions
columns_to_keep = [
    # Demographics & Academics
    'age', 'gender', 'race_ethnicity', 'degree_program',
    
    # Q1: How they heard about CCM
    'heard_ccm_website', 'heard_social_media', 'heard_community_event', 
    'heard_family_friend', 'heard_current_student', 'heard_alumni', 
    'heard_hs_teacher', 'heard_hs_counselor', 'heard_in_app_ad', 
    'heard_employer', 'heard_billboard', 'heard_tv', 'heard_radio', 'heard_other',
    
    # Q2: Enrollment Drivers
    'impact_affordable_cost', 'impact_location', 'impact_program_choice',
    'impact_online_offerings', 'impact_family_friend_referral', 'impact_faculty_staff',
    'impact_reputation', 'impact_financial_aid', 'impact_scholarships',
    'impact_small_classes', 'impact_extracurricular', 'impact_transfer_credits',
    'impact_negative_experience', 'impact_nj_stars', 'impact_transfer_ability', 'impact_hs_credit',
    
    # Q3: Pre-College Events
    'event_open_house', 'event_instant_decision', 'event_on_campus_session',
    'event_titans_tuesday', 'event_women_who_dare', 'event_college_fair',
    'event_sharetime', 'event_challenger', 'event_cybersecurity',
    'event_hs_info_session', 'event_hs_campus_visit', 'event_individual_campus_visit',
    'event_workforce_dev', 'event_corporate_training', 'event_teen_arts', 'event_summer_camp',
    
    # Q4: Prior Computing Exposure
    'computing_hs_class', 'computing_hs_club', 'computing_competitions',
    'computing_afterschool', 'computing_summer_camp', 'computing_ap_class',
    'computing_dual_enrollment', 'computing_family_influence', 'computing_family_in_field',
    'computing_hs_teacher_counselor', 'computing_employer_influence', 'computing_current_student',
    'computing_work_experience', 'computing_other',
    
    # Q5: Motivations for Computing Degree
    'motivation_job', 'motivation_transfer_bach', 'motivation_hs_transfer',
    'motivation_career_advancement', 'motivation_career_change', 'motivation_prof_development',
    'motivation_job_displacement', 'motivation_relocation', 'motivation_stay_current',
    'motivation_certifications', 'motivation_financial', 'motivation_personal_enrichment'
]

# Create a new cleaned dataframe with only the relevant columns
df_cleaned = df[columns_to_keep].copy()

# Verify the shape to ensure columns were successfully dropped
df_cleaned.shape
```

```python
# condense different majors and clean up messy text entries into unified categories
majors_cleanup = {
    # original tracks
    'Computer Science - Game Development': 'Computer Science',
    'Computer Science - Traditional': 'Computer Science',
    'Information Technology - Web Dev': 'Information Technology',
    'Information Technology - Networking': 'Information Technology',
    'CIS Game Development Option': 'Computer Science',
    'ShareTime CSIP Program': 'Computer Science',

    # business tracks
    'business': 'Business',
    'Business Admin': 'Business',
    'Business Administration': 'Business',
    'business administration': 'Business',
    
    # engineering
    'Mech e': 'Mechanical Engineering',
    'Mech E': 'Mechanical Engineering',
    'MET': 'Mechanical Engineering',
    'mechanical engineering': 'Mechanical Engineering',
    'Mechanical Engineering Technology': 'Mechanical Engineering',
    'Mechanical engineering technology': 'Mechanical Engineering',
    'Mechanical Engineering Tech': 'Mechanical Engineering',
    'Mechanical Engineering Technologist': 'Mechanical Engineering',
    'Mechanical Engineering Technologies': 'Mechanical Engineering',
    'Engineering Science / Mechanical Engineering Technology': 'Mechanical Engineering',
    'Mechanical Technology': 'Mechanical Engineering',
    'electrical engineering': 'Electrical Engineering',
    'electrical engineering technology': 'Electrical Engineering',
    'Electrical Engineering w/ biomed option': 'Electrical Engineering',
    'Electronical Engineer': 'Electrical Engineering',
    'Electronics Technolgies Engineering': 'Electrical Engineering',
    
    # cybersecurity and data science
    'cybersecurity': 'Cybersecurity',
    'cyberseucirty': 'Cybersecurity', 
    'data science': 'Data Science',
    'Computer Science (Data Science Option)': 'Data Science'
}

df_cleaned['degree_program'] = df_cleaned['degree_program'].replace(majors_cleanup)
df_cleaned['degree_program'].value_counts()
```

```python
# finding rows where the student selected more than one race
multiple_races = df_cleaned['race_ethnicity'].str.contains(';',na=False)

# replace specific rows with 'multi-racial'
df_cleaned.loc[multiple_races, 'race_ethnicity'] = 'Multi-Racial'

# verifying the changes
df_cleaned['race_ethnicity'].value_counts()
```

```python
# exporting data to a csv file
df_cleaned.to_csv('./cleaned_majors_survey_results.csv', index=False)
```
